```python
import pandas as pd
```

```python
df = pd.read_csv('output.csv')
```

```python
df.columns
```

```python
df.info()
```

```python
df = df.drop(labels='Unnamed: 0',axis=1)
```

```python
pd.set_option('display.max_columns',200)
pd.set_option('display.max_rows',200)
```

```python
df.describe()
```

```python
df1 = df[['Asystent parkowania','Marka pojazdu','Moc','Pojemność skokowa','Przebieg','Rodzaj paliwa','Rok produkcji','Skrzynia biegów','Cena']]
```

```python
df1 = df1.rename(columns={'Asystent parkowania':'asystent_parkowania',
                   'Marka pojazdu':'marka',
                    'Moc':'moc',
                    'Pojemność skokowa':'pojemnosc',
                    'Przebieg':'przebieg',
                    'Rodzaj paliwa':'rodzaj_paliwa',
                    'Rok produkcji':'rok_produkcji',
                    'Skrzynia biegów':'skrzynia_biegow',
                    'Cena':'cena'
                   })
```

```python
df1['asystent_parkowania'] = df1['asystent_parkowania'].fillna(0)
```

```python
list_ = (df1['marka'].value_counts() >= 10)
list_2 = list_.loc[list_ == True]
list_3 = pd.DataFrame(list_2).reset_index()['index']
list_4  =list(list_3)
```

```python
df2 = df1[df1['marka'].isin(list_4)]
```

```python
df2['marka'] = pd.Categorical(df2['marka'])
```

```python
df2['marka_cat'] = df2.marka.cat.codes
```

```python
df2
```

```python
df2 = df2.dropna()
```

```python
moc = []
for i,item in df2.iterrows():
    val = df2['moc'][i]
    y = val.split(' ')[0]
    moc.append(y)
```

```python
df2['moc_km'] = moc
```

```python
skrzynia = pd.get_dummies(df2['skrzynia_biegow'], prefix='skrzynia')[['skrzynia_Automatyczna']]
```

```python
df3 = pd.concat([df2,skrzynia],axis=1)
```

```python
pojemnosc = []
for i,item in df3.iterrows():
    y = ''.join(df3['pojemnosc'][i]).split(' ')[:-1]
    string = ''.join([str(item) for item in y])
    pojemnosc.append(int(string))
```

```python
df3['pojemnosc_skokowa [cm3]'] = pojemnosc
```

```python
df3['rok_produkcji'] = df3['rok_produkcji'].astype('int')
```

```python
przebieg = []
for i,item in df3.iterrows():
    y = ''.join(df3['przebieg'][i]).split(' ')[:-1]
    string = ''.join([str(item) for item in y])
    przebieg.append(int(string))
```

```python
df3['przebieg [km]'] = przebieg
```

```python
paliwo = pd.get_dummies(df3['rodzaj_paliwa'])
```

```python
df4 = pd.concat([df3,paliwo],axis=1)
```

```python
df4
```

```python
df_mdl = df4[['rok_produkcji','asystent_parkowania','marka_cat','moc_km','skrzynia_Automatyczna','pojemnosc_skokowa [cm3]','przebieg [km]','Benzyna','Benzyna+CNG','Benzyna+LPG','Diesel','Hybryda','cena']]
```

```python
import numpy as np
```

```python
nowy_x = []
for i,item in df_mdl.iterrows():
    try:
        x = df_mdl['cena'][i].split(',')[0]
        nowy_x.append(int(x))
    except:
        nowy_x.append(int(df_mdl['cena'][i]))
```

```python
df_mdl['cena'] = nowy_x
```

```python
pd.set_option('display.max_rows',300)
```

```python
X = df_mdl.drop(labels='cena',axis=1)
```

```python
y = df_mdl[['cena']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
```

```python
mdl = RandomForestRegressor(n_estimators = 70)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
mdl.fit(X_train, y_train)
```

```python
predictions = mdl.predict(X_test)
```

```python
mean_squared_error(y_test,predictions)
```

```python
r2_score(y_test,predictions)
```

```python
mean_absolute_error(y_test,predictions)
```
